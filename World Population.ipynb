# Import Python Libraries

1- Scientific Computing Libraries:
   * Pandas: offers -data structures &tools- for effective data manipulation, analaysis, cleaning, and exploring ((dataframe))
   * NumPy: used for working with Arrays & Matrices
   * SciPy: Integrals, Solving differential Equations, Optimization, Advanced Math problems, as well as data Visualization
   
2- Visualization Libraries in Python:
   * Matplotlib: plots &Graphs
   * Seaborn: plots: Heatmap, Timeseries, Violin plots 
   
3- Algorithmic Libraries in Python:
   * Scikit-learn: Machine learning (Regression, Classification, Clustering, ...)
   * Statsmodels: Explore data, Estimate Statistical Models, and Perform Statistical tests

```python
#### show populations only to visualize it ####

#df2 = df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by = '2022 Population', ascending = False)


df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by = '2022 Population', ascending = False)
df2
```

```python
#### transpose: Columns will be indexes, and continents will be columns ####

df3 = df2.transpose()
df3
```

```python
#### Line Plot ####
#df3.plot(kind = 'line')

df3.plot()
```

```python
######################### Bar Plot (Bar Chart) #########################
#### the default is to plot it vertically but we can plot it horizontally ####

df3.plot(kind = 'bar')
```

```python
######################### Stacked Bar Plot (Bar Chart) #########################
#### the default is to plot it vertically but we can plot it horizontally ####

df3.plot(kind = 'bar', stacked = True)
```
